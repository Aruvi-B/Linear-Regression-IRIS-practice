```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 42)
```

```python
# Import Libraries
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_scaled, y_train.values.ravel())
rf_predition = rf_model.predict(X_test_scaled)
```

```python
# Naive Bayes
nb_model = GaussianNB()
nb_model.fit(X_train_scaled, y_train.values.ravel())
nb_predition = nb_model.predict(X_test_scaled)
```

```python
# Neural Network
nn_model = MLPClassifier(random_state=42)
nn_model.fit(X_train_scaled, y_train.values.ravel())
nn_predition = nn_model.predict(X_test_scaled)
```

```python
# SVM
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train_scaled, y_train.values.ravel())
svm_predition = svm_model.predict(X_test_scaled)
```

```python
# Gradient Boosting
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train_scaled, y_train.values.ravel())
gb_predition = gb_model.predict(X_test_scaled)
```

```python
# K-Nearest Neighbors
knn_model = KNeighborsClassifier()
knn_model.fit(X_train_scaled, y_train.values.ravel())
knn_predition = knn_model.predict(X_test_scaled)
```

```python
models = ['Random Forest', 'Naive Bayes',"SVM", "Gradient Boosting", "KNN", 'Neural Network']
predictions = [rf_predition,  nb_predition,  svm_predition, gb_predition, knn_predition, nn_predition]

for model, pred in zip(models, predictions):
    accuracy = accuracy_score(y_test, pred)
    conf_matrix = confusion_matrix(y_test, pred)
    classification_rep = classification_report(y_test, pred)

    print(f"Model: {model}")
    print(f"Accuracy: {accuracy}")
    print("Confusion Matrix:\n", conf_matrix)
    print("Classification Report:\n", classification_rep)

    print("__________________________________________________________________")


    # confusion matrix
    plt.figure(figsize = (8, 6))
    sns.heatmap(conf_matrix,annot = True, fmt = 'd', linewidth = 0.5, square = True)
    plt.show()
```
