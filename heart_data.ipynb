```python
df_train = df
from sklearn.neighbors import KNeighborsClassifier

X = df.drop('cardio', axis=1)
y = df['cardio']
```

```python
scaler = MinMaxScaler()
X_scale = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scale,y,train_size=0.8)
print("Total", len(X))
print("Train size ", len(X_train))
print("Test size ", len(X_test))
```

```python
# XG Boost
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
```

```python
print("XGBoost Accuracy: ", accuracy_score(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))
```

```python
# prompt: 𝐾-Nearest Neighbor
# 2. Logistic Regression
# 3. Decision Tree
# 4. Random Forest
# i need more hyper tuning and give me the best result

# K-Nearest Neighbor
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
print("KNN Accuracy: ", accuracy_score(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))

# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=0.1, solver='liblinear')
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
print("Logistic Regression Accuracy: ", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))

# Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='entropy', random_state=0, max_depth=6)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
print("Decision Tree Accuracy: ", accuracy_score(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))

# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=0)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Random Forest Accuracy: ", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))

```

```python
# Hyperparameter tuning
from sklearn.model_selection import GridSearchCV

# KNN
param_grid_knn = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}
grid_knn = GridSearchCV(knn, param_grid_knn, cv=5)
grid_knn.fit(X_train, y_train)
print("Best KNN parameters:", grid_knn.best_params_)

```

```python
# Logistic Regression
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear', 'sag', 'saga'],
    'penalty': ['l1', 'l2', 'elasticnet', 'none']
}
grid_lr = GridSearchCV(lr, param_grid_lr, cv=5)
grid_lr.fit(X_train, y_train)
print("Best Logistic Regression parameters:", grid_lr.best_params_)

```

```python
# Decision Tree
param_grid_dt = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 7, 9],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_dt = GridSearchCV(dt, param_grid_dt, cv=5)
grid_dt.fit(X_train, y_train)
print("Best Decision Tree parameters:", grid_dt.best_params_)

```

```python
# Random Forest
param_grid_rf = {
    'n_estimators': [50, 100, 150],
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 7, 9],
    # 'min_samples_split': [2, 5, 10],
    # 'min_samples_leaf': [1, 2, 4]
}
grid_rf = GridSearchCV(rf, param_grid_rf, cv=5)
grid_rf.fit(X_train, y_train)
print("Best Random Forest parameters:", grid_rf.best_params_)

```

```python
# Train and evaluate the best models
knn_best = KNeighborsClassifier(**grid_knn.best_params_)
lr_best = LogisticRegression(**grid_lr.best_params_)
dt_best = DecisionTreeClassifier(**grid_dt.best_params_)
rf_best = RandomForestClassifier(**grid_rf.best_params_)

```

```python
knn_best.fit(X_train, y_train)
lr_best.fit(X_train, y_train)
dt_best.fit(X_train, y_train)
rf_best.fit(X_train, y_train)

```

```python
y_pred_knn_best = knn_best.predict(X_test)
y_pred_lr_best = lr_best.predict(X_test)
y_pred_dt_best = dt_best.predict(X_test)
y_pred_rf_best = rf_best.predict(X_test)

```

```python
print("Best KNN Accuracy:", accuracy_score(y_test, y_pred_knn_best))
print("Best Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr_best))
print("Best Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt_best))
print("Best Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf_best))
```
